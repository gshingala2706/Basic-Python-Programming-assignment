## 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
i =0

while(i < 3):
    
    num = int(input('Enter the number: '))

    if num == 0:
        print(f'Entered number {num}  is Zero')

    elif num > 0:
        print(f'Entered number {num} is Positive number')

    else:
        print(f'Entered number {num} is Negative')
        
    i+=1
```

## 2.	Write a Python Program to Check if a Number is Odd or Even?

```python
num = int(input('Enter the number: '))

if num%2 == 0:
    print(f'{num} Number is Even')
    
else:
    print(f'{num} Number is Odd')
```

## 3.	Write a Python Program to Check Leap Year?

```python
year = int(input('Enter the year: '))

if year%4 == 0:
    if year%100 != 0:
        print(f'{year} Year is leap year')
        
    elif year%100 == 0:
        if year%400 == 0:
            print(f'{year} Year is leap year')
            
        else:
            print(f'{year} Year is not leap year')
                      
else:
    print(f'{year} Year is not leap year')
```

## 4.	Write a Python Program to Check Prime Number?

```python
# Prime number: number is exactly divisible by itself or by 1
# Composite number: Except prime number, remaining all number are composite number

num = int(input('Enter the number: '))

if num == 1 or num == 0:
    print(f'{num} neither prime nor composite number')
    
else:
    for i in range(2, num):
        if num%i == 0:
            print(f'{num} Number is not a prime number (composite number)')
            print(i,"times",num//i,"is",num)
            break
            
            
    else:
        print(f'{num} Number is a prime number')
    
```
